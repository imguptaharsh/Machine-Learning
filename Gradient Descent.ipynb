# Gradient  Descent( in general)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
y=(x-5)**2
print(x,y)
```

**Goal**
*Given a function f(x), we want to find the value of x that minimum*
## Visualisation

```python
plt.plot(x,y)
plt.ylabel('F(x)')
plt.xlabel('x')
plt.show()
```

```python
x1= 0
lr=0.1
error= []
# 50 steps in the downhill direction
plt.plot(x,y)
for i in range(50):
    grad= 2*(x1-5)
    x1= x1- lr*grad
    e = (x1-5)**2
    error.append(e)
    plt.scatter(x1,e)
    plt.show
#     print(x1)
```

```python
plt.plot(error)
plt.show()
```
